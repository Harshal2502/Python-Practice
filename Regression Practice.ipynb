```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([1,2,3,4, 5, 6, 7, 9])
y = np.array([2,4,3,5, 4, 4, 6, 5])
tempX = pd.DataFrame({'col':X})
tempy = pd.DataFrame({'col':y})
```

```python
plt.figure(figsize=(10,6))
plt.grid()
plt.plot(X, y, color='red')
plt.show()
```

```python
regression = LinearRegression()
regression.fit(tempX, tempy)
```

```python
regression.coef_[0][0] 
```

```python
regression.intercept_[0]
```

```python
regression.score(tempX, tempy)
```

```python
plt.figure(figsize=(10,6))
plt.grid()
plt.scatter(X, y, color='red')
plt.plot(tempX['col'], regression.predict(tempX), linewidth=2)
plt.show()
```
